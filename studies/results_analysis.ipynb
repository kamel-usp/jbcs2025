```python
import pandas as pd
import glob
import os

# Path to the root folder containing the date/time subfolders
root_folder = "/home/andrebarbosa/jbcs2025/outputs"

# Use glob to find all 'evaluation_results.csv' files in subfolders
all_csv_paths = glob.glob(os.path.join(root_folder, "**", "evaluation_results.csv"), recursive=True)
all_csv_paths
```

```python
df_list = []
for csv_path in all_csv_paths:
    # Read the CSV
    temp_df = pd.read_csv(csv_path)
    temp_df.columns = temp_df.columns.str.replace("eval_","")
    # Append to our list
    df_list.append(temp_df)

# Combine all into a single DataFrame
combined_df = pd.concat(df_list, ignore_index=True)

# Now you have one DataFrame with all rows from each evaluation_results.csv
combined_df[~combined_df["step"].isin(["baseline_evaluation","evaluation_after_training"])].sort_values(by="timestamp")
```
